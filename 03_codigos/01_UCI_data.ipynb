```python
# Importando bibliotecas
from ucimlrepo import fetch_ucirepo
import os
from datetime import datetime 

# Definindo paths
path_codigos = os.getcwd()
path_projeto = os.path.dirname(path_codigos)
path_data = os.path.join(path_projeto, "02_data")
path_bronze = os.path.join(path_data, "01_bronze")
path_silver = os.path.join(path_data, "02_silver")
```

```python
# Importando dados
adult = fetch_ucirepo(id = 2)

# Capturando dataset
adult_dataset = adult.data.original

del adult

# Salvando na bronze layer
adult_dataset.to_parquet(os.path.join(path_bronze, "adult_dataset.parquet"), index = False)

# Criando coluna de adult_id e event_timestamp
adult_dataset = adult_dataset.reset_index()
adult_dataset = adult_dataset.reset_index()
adult_dataset = adult_dataset.rename(columns = {"index": "adult_id"})
adult_dataset = adult_dataset.rename(columns = {"level_0": "event_timestamp"})

# Populando coluna de timestamp
adult_dataset["event_timestamp"] = datetime(2022,10,1)
adult_dataset.loc[(10000 < adult_dataset["adult_id"]) & (adult_dataset["adult_id"] <= 20000), "event_timestamp"] = datetime(2023,3,1)
adult_dataset.loc[(20000 < adult_dataset["adult_id"]) & (adult_dataset["adult_id"] <= 30000), "event_timestamp"] = datetime(2023,10,1)
adult_dataset.loc[(30000 < adult_dataset["adult_id"]) & (adult_dataset["adult_id"] <= 48000), "event_timestamp"] = datetime(2024,3,1)
adult_dataset.loc[48000 < adult_dataset["adult_id"], "event_timestamp"] = datetime(2024,10,1)

# Dividindo dados em dados históricos e dados de inferência
adult_dataset_hist = adult_dataset.loc[adult_dataset["adult_id"] <= 48000].copy().reset_index(drop = True)
adult_dataset_inf = adult_dataset.loc[48000 < adult_dataset["adult_id"]].copy().reset_index(drop = True).drop(columns = ["income"])

# Salvando na silver layer
adult_dataset_hist.to_parquet(os.path.join(path_silver, "adult_dataset_hist.parquet"), index = False)
adult_dataset_inf.to_parquet(os.path.join(path_silver, "adult_dataset_inf.parquet"), index = False)

adult_dataset_hist.info()
display(adult_dataset_inf)

adult_dataset_inf.info()
display(adult_dataset_inf)
```
